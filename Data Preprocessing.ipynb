```python
import pandas as pd
import json
```

```python
with open('WPl_json_files/match_summary.json') as f:
    data =  json.load(f)
df_matches = pd.DataFrame(data)
df_matches.head()
```

```python
df_matches.to_csv('WPL_csv_files/match_summary.csv', index = False)
```

```python
match_dict = {}

for index,row in df_matches.iterrows():
    key1 = row['team1'] + ' Vs ' +row['team2']
    key2 = row['team2'] + ' Vs ' +row['team1']
    match_dict[key1] = row["matchID"]
    match_dict[key2] = row["matchID"]
print(match_dict)
```

**BATTING SUMMARY**

```python
with open('WPl_json_files/batting_summary.json') as f:
    data =  json.load(f)
    batting_records = []
    for rec in data:
        batting_records.extend(rec)
        
df_batting = pd.DataFrame(batting_records)
```

```python
df_batting["out/not_out"] = df_batting.dismissal.apply(lambda x: "out" if len(x)>1 else "not_out")
df_batting.drop(columns=["dismissal"],inplace=True)
df_batting.head()
```

```python
df_batting['match'] = df_batting['match'].apply(lambda x: x.rstrip())
df_batting
```

```python
df_batting["matchID"] = df_batting["match"].map(match_dict)
df_batting.head()
```

```python
df_batting.to_csv('WPL_csv_files/batting_summary.csv', index = False)
```
